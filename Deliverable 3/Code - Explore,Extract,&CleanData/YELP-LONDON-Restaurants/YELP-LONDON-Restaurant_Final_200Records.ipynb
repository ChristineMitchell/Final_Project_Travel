```python
# Dependencies and Setup
import pandas as pd
import requests
import json

# Import API key
from config import api_key
```

Yelp Business Search API Documentation: https://www.yelp.com/developers/documentation/v3/business_search

```python
# headers contain the api key.
# reference this site for set up: https://www.justintodata.com/python-api-call-to-request-data/
headers = {"Authorization": "Bearer {}".format(api_key)}
search_api_url = "https://api.yelp.com/v3/businesses/search"
```

Yelp Business Search API default returns twenty records at a time. However, the API allows fifty records per call by adding a limit to the parameters, with four calls maximum before returning an error. So, the maximum dataset size is 200 records.

```python
#First set of 50 records 0-50
lndr_params1 = {"term":"restaurant","location":"London","limit":50,"offset":0}
lndr_response1 = requests.get(search_api_url, headers=headers, params=lndr_params1)
lndr1 = lndr_response1.json()
display(lndr1)
#print(lndr_response1.url)
#print(lndr_response1.status_code)
```

```python
#Create a for loop to capture the data of the first 50 records 0-50

#Create Empty Lists
lat1=[]
lng1=[]
address1=[]
name1=[]
title1=[]
price1=[]
rating1=[]
ids1=[]
url1=[]

#Set Counter to 0
i = 0

#For Loop with Try/Except Error
for i in (lndr1["businesses"]):
    try:
        lat1.append (i["coordinates"]["latitude"])
        lng1.append (i["coordinates"]["longitude"])
        address1.append (i["location"]["address1"])
        name1.append (i["name"])    
        title1.append(i["categories"][0]["title"])
        price1.append (i["price"])
        rating1.append (i["rating"])
        ids1.append (i["id"])
        url1.append (i["url"])
    except(KeyError,IndexError):
        print("Skipping, Not Found")
        price1.append("")
        rating1.append("")
        ids1.append("")
        url1.append("")
```

```python
#Create DataFrame of the first 50 records collected 
df1 = pd.DataFrame({"Restaurant": name1,"Type": title1, "Price":price1, "Rating":rating1, "Address":address1, "Lat":lat1, "Lng":lng1, "url":url1, "ID":ids1 })
df1.head()
```
