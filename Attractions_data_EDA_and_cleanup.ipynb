```python
#cleaning
df = pd.DataFrame(art_gallery_df['the_geom'])
df[['A', 'B']] = df['the_geom'].str.split(' ', 1, expand=True)
art_gallery_df.count()
```

```python
gallerys_df = pd.DataFrame(columns=["Name", "Geometry", "Address", "Description", "Lattitude", "Longitude", "Category"])
gallerys_df["Name"] = art_gallery_df["NAME"]
gallerys_df["Address"] = art_gallery_df["ADDRESS1"]
gallerys_df["Category"] = "Art_Gallery"
gallerys_df["Geometry"] = art_gallery_df["the_geom"]
gallerys_df[['Lattitude', 'Longitude']] = df['B'].str.split(' ', 1, expand=True)
gallerys_df['Lattitude'] = gallerys_df['Lattitude'].map(lambda x: x.lstrip('(').rstrip(')'))
gallerys_df['Longitude'] = gallerys_df['Longitude'].map(lambda x: x.lstrip('(').rstrip(')'))
gallerys_df
```

```python
df = pd.DataFrame(museum_df['the_geom'])
df[['A', 'B']] = df['the_geom'].str.split(' ', 1, expand=True)
museum_df.count()
```

```python
museums_df = pd.DataFrame(columns=["Name", "Geometry", "Address", "Description", "Lattitude", "Longitude", "Category"])
museums_df["Name"] = museum_df["NAME"]
museums_df["Address"] = museum_df["ADRESS1"]
museums_df["Category"] = "Museum"
museums_df["Geometry"] = museum_df["the_geom"]
museums_df[['Lattitude', 'Longitude']] = df['B'].str.split(' ', 1, expand=True)
#museums_df['Longitude'] = museums_df[museums_df['Longitude'].str.replace(")", "")]
museums_df['Lattitude'] = museums_df['Lattitude'].map(lambda x: x.lstrip('(').rstrip(')'))
museums_df['Longitude'] = museums_df['Longitude'].map(lambda x: x.lstrip('(').rstrip(')'))
museums_df
```

```python
park_monuments_df.count()
```

```python
monuments_filtered = park_monuments_df[park_monuments_df['X'].notnull()]
monuments_filtered = monuments_filtered[monuments_filtered['parkname'].notnull()]
monuments_filtered.count()
```

```python
def latlang_to_geodata(lat, long):
    s = pd.DataFrame()
    for i in range(len(lat)):
        s.append = "POINT(" + str(lat) + " " + str(long) + ")"
    return s

parks_df = pd.DataFrame(columns=["Name", "Geometry", "Address", "Description", "Lattitude", "Longitude", "Category"])
parks_df["Name"] = monuments_filtered["name"]
#for i in range(len(parks_df)):.append
parks_df["Geometry"] = "POINT(" + monuments_filtered["X"].astype(str) + " " + monuments_filtered["Y"].astype(str) + ")"
#parks_df["Geometry"] = latlang_to_geodata(monuments_filtered["X"], monuments_filtered["Y"])
parks_df["Address"] = monuments_filtered["Location"]
parks_df["Category"] = "Park"
parks_df["Lattitude"] = monuments_filtered["X"]
parks_df["Longitude"] = monuments_filtered["Y"]
parks_df
```

```python
kayaking_stuff_df = water_df[water_df['lat'].notnull()]
kayaking_stuff_df.count()
```

```python
kayaking_df = pd.DataFrame(columns=["Name", "Geometry", "Address", "Description", "Lattitude", "Longitude", "Category"])
kayaking_df["Name"] = kayaking_stuff_df["Name"]
kayaking_df["Address"] = kayaking_stuff_df["Location"]
kayaking_df["Geometry"] = "POINT(" + kayaking_stuff_df["lat"].astype(str) + " " + kayaking_stuff_df["lon"].astype(str) + ")"
kayaking_df["Category"] = "Kayaking/Canoeing"
kayaking_df["Lattitude"] = kayaking_stuff_df["lat"]
kayaking_df["Longitude"] = kayaking_stuff_df["lon"]
kayaking_df
```

```python
grafiting_df = grafiti_df[grafiti_df['lat'].notnull()]
grafiting_df.count()
```

```python
streetart_df = pd.DataFrame(columns=["Name", "Geometry", "Address", "Description", "Lattitude", "Longitude", "Category"])
streetart_df["Name"] = grafiting_df["name"]
streetart_df["Address"] = grafiting_df["location"]
streetart_df["Geometry"] = "POINT(" + grafiting_df["lat"].astype(str) + " " + grafiting_df["lng"].astype(str) + ")"
streetart_df["Category"] = "StreetArt"
streetart_df["Lattitude"] = grafiting_df["lat"]
streetart_df["Longitude"] = grafiting_df["lng"]
streetart_df
```

```python
new_york_df = pd.DataFrame(columns=["Name", "Geometry", "Address", "Description", "Lattitude", "Longitude", "Category"])
new_york_df = new_york_df.append(gallerys_df, ignore_index=True)
new_york_df = new_york_df.append(museums_df, ignore_index=True)
new_york_df = new_york_df.append(parks_df, ignore_index=True)
new_york_df = new_york_df.append(kayaking_df, ignore_index=True)
new_york_df = new_york_df.append(streetart_df, ignore_index=True)
new_york_df
```

```python
new_york_df.to_csv("Resources/NewYorkTotal.csv")
```
