```python
not_wanted = ["Dentist's Office", 'Airport Gate', 'Airport Terminal', 'Home (private)',
             'Office', 'Coworking Space', 'Military Base', 'Residential Building (Apartment / Condo)', 'Eye Doctor', 
              'Voting Booth', 'Conference Room', 'Veterinarian', 'Fire Station', 'Police Station', 'Post Office', 
              'Professional & Other Places', 'College Science Building', 'Hotel', 'College Academic Building', 
             'School', 'General College & University']
airport = ['Airport Lounge', 'Airport Tram', 'Airport Food Court']
medical = ['Medical Center', "Doctor's Office", 'Hospital', 'Emergency Room', ]
misc = ['Grocery Store', 'Cosmetics Shop', 'Gaming Cafe', 'Coffee Shop', 'Mall', 'Plaza']
water = ['Pier', 'Harbor / Marina']
history = ['History Museum', 'Historic Site', 'Museum']
lib = ['College Library']
```

```python
london_poi_df.drop(london_poi_df.loc[london_poi_df.Subcategory.isin(not_wanted)].index, inplace=True)
london_poi_df.loc[london_poi_df.Subcategory.isin(airport)] = 'Airport'
london_poi_df.loc[london_poi_df.Subcategory.isin(medical)] = 'Medical Building'
london_poi_df.loc[london_poi_df.Subcategory.isin(misc)] = 'Miscellaneous Shop'
london_poi_df.loc[london_poi_df.Subcategory.isin(water)] = 'Pier / Marina'
london_poi_df.loc[london_poi_df.Subcategory.isin(history)] = 'Museum / Historic Site'
london_poi_df.loc[london_poi_df.Subcategory.isin(lib)] = 'Library'
```

```python
poi_df = pd.DataFrame(columns=["Name", "Geometry", "Address", "Lattitude", "Longitude", "Category"])
poi_df["Name"] = london_poi_df["Name"]
poi_df["Geometry"] = "POINT(" + london_poi_df["Lat"].astype(str) + " " + london_poi_df["Lng"].astype(str) + ")"
poi_df["Address"] = london_poi_df["Address"]
poi_df["Category"] = london_poi_df["Subcategory"]
poi_df["Lattitude"] = london_poi_df["Lat"]
poi_df["Longitude"] = london_poi_df["Lng"]
poi_df
```

```python
attract_df = pd.DataFrame(columns=["Name", "Geometry", "Address", "Lattitude", "Longitude", "Category"])
attract_df["Name"] = london_attractions_df["Name"]
attract_df["Geometry"] = "POINT(" + london_attractions_df["Lat"].astype(str) + " " + london_attractions_df["Lng"].astype(str) + ")"
attract_df["Address"] = london_attractions_df["Address"]
attract_df["Category"] = london_attractions_df["Category"]
attract_df["Lattitude"] = london_attractions_df["Lat"]
attract_df["Longitude"] = london_attractions_df["Lng"]
attract_df
```

```python
london_df = pd.DataFrame(columns=["Name", "Geometry", "Address", "Lattitude", "Longitude", "Category"])
london_df = london_df.append(attract_df, ignore_index=True)
london_df = london_df.append(poi_df, ignore_index=True)
london_df
```
