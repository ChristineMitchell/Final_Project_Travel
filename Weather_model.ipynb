```python
# Importing the dependencies
import pandas as pd
import psycopg2
from getpass import getpass
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
```

```python
# Cleaning the data
# Dropping the columns that aren't needed
london_weather_df = london_weather_df.drop(columns=['index'], axis=1)
nyc_weather_df = nyc_weather_df.drop(columns=['index'], axis=1)
```

```python
# Checking for null values
print(london_weather_df.isnull().sum())
print(nyc_weather_df.isnull().sum())
```

```python
# Checking our datatypes
print(london_weather_df.dtypes)
print(nyc_weather_df.dtypes)
```

```python
# Splitting the features
london_X = london_weather_df.drop(columns=['avg_high_temp_f', 'year', 'month_num'], axis=1)
nyc_X = nyc_weather_df.drop(columns=['avg_high_temp_f', 'year', 'month_num'], axis=1)
london_y = london_weather_df['avg_high_temp_f']
nyc_y = nyc_weather_df['avg_high_temp_f']
```

```python
# Splitting the data into training and testing datasets
london_X_train, london_X_test, london_y_train, london_y_test = train_test_split(london_X, london_y)
nyc_X_train, nyc_X_test, nyc_y_train, nyc_y_test = train_test_split(nyc_X, nyc_y)
```

```python
# Looking at the shapes of our train and test sets
print(london_X_train.shape)
print(london_X_test.shape)
print(london_y_train.shape)
print(london_y_test.shape)
print('---')
print(nyc_X_train.shape)
print(nyc_X_test.shape)
print(nyc_y_train.shape)
print(nyc_y_test.shape)
```

```python
# Scaling the London data
# Creating a StandardScaler instance
scaler = StandardScaler()

# Training the scaler
london_X_scaler = scaler.fit(london_X_train)

# Scaling the data
london_X_train_scaled = london_X_scaler.transform(london_X_train)
london_X_test_scaled = london_X_scaler.transform(london_X_test)
```

```python
# Scaling the NYC data
# Creating a StandardScaler instance
scaler = StandardScaler()

# Training the scaler
nyc_X_scaler = scaler.fit(nyc_X_train)

# Scaling the data
nyc_X_train_scaled = nyc_X_scaler.transform(nyc_X_train)
nyc_X_test_scaled = nyc_X_scaler.transform(nyc_X_test)
```

```python
# Defining the models
london_num_input_features = len(london_X_train_scaled[0])
london_hidden_nodes_layer_1 = 20
london_hidden_nodes_layer_2 = 10
london_hidden_nodes_layer_3 = 5

nyc_num_input_features = len(nyc_X_train_scaled[0])
nyc_hidden_nodes_layer_1 = 20
nyc_hidden_nodes_layer_2 = 10
nyc_hidden_nodes_layer_3 = 5
```

```python
# Creating the London model
# Creating the neural network instance
london_model = tf.keras.models.Sequential()

# Creating the London model
# First hidden layer
london_model.add(tf.keras.layers.Dense(units=london_hidden_nodes_layer_1, 
                                        input_dim=london_num_input_features, activation='relu'))
# Second hidden layer
london_model.add(tf.keras.layers.Dense(units=london_hidden_nodes_layer_2, activation='relu'))
# Third hidden layer
london_model.add(tf.keras.layers.Dense(units=london_hidden_nodes_layer_3, activation='relu'))
# Output layer
london_model.add(tf.keras.layers.Dense(units=1, activation='linear'))

# Checking the structure of the model
london_model.summary()
```

```python
# Creating the NYC model
# Creating the neural network instance
nyc_model = tf.keras.models.Sequential()

# Creating the NYC model
# First hidden layer
nyc_model.add(tf.keras.layers.Dense(units=nyc_hidden_nodes_layer_1, 
                                        input_dim=nyc_num_input_features, activation='relu'))
# Second hidden layer
nyc_model.add(tf.keras.layers.Dense(units=nyc_hidden_nodes_layer_2, activation='relu'))
#Third hidden layer
nyc_model.add(tf.keras.layers.Dense(units=nyc_hidden_nodes_layer_3, activation='relu'))
# Output layer
nyc_model.add(tf.keras.layers.Dense(units=1, activation='linear'))

# Checking the structure of the model
nyc_model.summary()
```

```python
# Compiling the models
london_model.compile(loss='mse', optimizer='adam', metrics=['MSE'])
nyc_model.compile(loss='mse', optimizer='adam', metrics=['MSE'])
```

```python
# Training the London model
london_fit_model = london_model.fit(london_X_train_scaled, london_y_train, epochs=500)
```

```python
# Training the NYC model
nyc_fit_model = nyc_model.fit(nyc_X_train_scaled, nyc_y_train, epochs=500)
```

```python
# Evaluating the models
london_model_loss, london_model_mse = london_model.evaluate(london_X_test_scaled, london_y_test)
nyc_model_loss, nyc_model_mse = nyc_model.evaluate(nyc_X_test_scaled, nyc_y_test)

print(f'London Loss:  {london_model_loss}, London MSE: {london_model_mse}')
print(f'NYC Loss:  {nyc_model_loss}, NYC MSE: {nyc_model_mse}')
```
